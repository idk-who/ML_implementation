```python
import time
import numpy as np
import torch
from torch import nn
from d2l import torch as d2l
```

```python
class Temp:
    def __init__(self):
        self.a = 16

A = Temp()
A.a
```

```python
def add_to_class(Class):
    
    def wrapper(func):
        setattr(Class, func.__name__, func)
    
    return wrapper
```

```python
@add_to_class(Temp)
def set_a(self, val):
    self.a = val
```

```python
A.set_a(64)
A.a
```

```python
class Temp2(d2l.HyperParameters):
    def __init__(self, a, b, not_to_save):
        self.save_hyperparameters(ignore = ['not_to_save'])
```

```python
obj2 = Temp2(1, 2, 9999)
```

```python
obj2.a, obj2.b
```
